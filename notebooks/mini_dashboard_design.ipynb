```python
import plotly.express as px
import pandas as pd
```

```python
df=pd.read_csv("reviews_products_with_sentiment.csv")
df.head()
```

Sentiment Distribution

Goal: Understand the overall balance between positive and negative reviews.
Plot: Bar chart or pie chart.

```python
fig1 = px.histogram(
    df,
    x="sentiment",
    color="sentiment",
    title="Sentiment Distribution"
)

fig1.update_layout(
    yaxis_title="Number of Customer Responses"
)

fig1.show()
```

Average Confidence per Sentiment

Goal: See how confident your model was when labeling each sentiment class.
Plot: Bar chart of mean confidence grouped by sentiment.

```python
avg_conf = df.groupby("sentiment", as_index=False)["confidence"].mean()

fig2 = px.bar(
    avg_conf,
    x="sentiment",
    y="confidence",
    color="sentiment",
    title="Model Confidence by Sentiment",
    color_discrete_sequence=["#EF553B", "#636EFA"]  # red first, blue second (inverted)
)

fig2.update_layout(
    yaxis_title="Average Model Confidence (%)"
)

fig2.show()
```

Ratings vs Sentiment

Goal: Compare human ratings (if your dataset has rating_rate) with model-predicted sentiment.
Plot: Boxplot or scatter plot.

```python
fig3 = px.box(df, x="sentiment", y="rating_rate",
              color="sentiment", title="Ratings vs Sentiment")
fig3
```

Sentiment by Product Category (or Price)

Goal: Identify which product types or price ranges get the most negative feedback.
Plot: Grouped bar chart.

```python
sent_by_cat = df.groupby(["category", "sentiment"]).size().reset_index(name="count")

fig4 = px.bar(
    sent_by_cat,
    x="category",
    y="count",
    color="sentiment",
    title="Customer Sentiment by Product Category",
    barmode="group",
    color_discrete_sequence=["#EF553B", "#636EFA"]  # NEGATIVE first, POSITIVE second
)

fig4.update_layout(
    yaxis_title="Number of Customer Responses",
    xaxis_title="Product Category"
)

fig4.show()
```
